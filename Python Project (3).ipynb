### Q1.Show the no.of Death Cases in Each State

```python
plt.figure(figsize = (10,7))
plt.xticks(rotation=90)
plt.xlabel('Deaths Cases',size=15)
plt.ylabel('Country/Region', size=15)
plt.title('Statewise Death Cases till today', color='red')
plt.bar(df['Country/Region'],df['Deaths'],width=10)
```

### Q2.Show the Active Cases in each country


```python
plt.figure(figsize = (10,7))
plt.xticks(rotation=90)
plt.xlabel('Recovery Rate',size=15)
plt.ylabel('Country/Region', size=15)
plt.title('Statewise Recovery Cases till today', color='red')
plt.plot(df['Country/Region'],df['Recovered'])
plt.show()
```

### Q3.Show the Confirmed Cases in each Country

```python
plt.figure(figsize = (10,7))
plt.xticks(rotation=90)
plt.xlabel('Confirmed Cases',size=15)
plt.ylabel('Country/Region', size=15)
plt.title('Statewise Confirmed Cases till today', color='red')
plt.scatter(df['Country/Region'],df['Confirmed'])
```

### Q4.Show the new cases in each country??

```python
x=df['Country/Region']
y=df['New cases']
plt.pie(y,labels=x)
plt.show
```
